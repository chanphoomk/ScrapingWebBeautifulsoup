```
#Checking weather we're allow to scrape or not
#If allow output should be 200

import requests 

from bs4 import BeautifulSoup 

result=requests.get("https://thestandard.co/homepage/")

print("website status : ",result.status_code )

print("Web Header : ",result.headers)
```

```
src = result.content
#To be able to pass HTML content
soup = BeautifulSoup(src,'lxml')
#print(soup.prettify())
```

```
#Queries 1 article

#Getting headlines news on THE STANDARD
article = soup.find("h2",class_='news-title')
headline = article.a.text
print(" First Headline :\n",headline, "\n")

#Getting link to the news (inside tag)
news_src = article.find('a')['href']
print(news_src, "\n")

#Getting word/sentence in the link above(SPLIT)
news_id = news_src.split('/')[3]
print("ID after website link is :",news_id,'\n')

#Make link to webpage from id
web_link = f'https://thestandard.co/{news_id}'
print("Your generated weblink : ", web_link, "\n")
```

```
#Queries All Article

for article in soup.find_all("h2",class_='news-title'):
  headline = article.a.text
  print("Today News :",headline)

  #Getting link to the news (inside tag)
  news_src = article.find('a')['href']
  print(news_src)

  #Getting word/sentence in the link above
  news_id = news_src.split('/')[3]
  print("ID after website link is :",news_id)

  #Make link to webpage from id
  web_link = f'https://thestandard.co/{news_id}'
  print("Your generated weblink : ", web_link, "\n")

  print()

```

```
#Way to fix missing data in the tag (TRY, EXECPT)

for article in soup.find_all("h2",class_='news-title'):
  headline = article.a.text
  print("Today News :",headline)
  try:
    #Getting link to the news (inside tag)
    news_src = article.find('a')['href']
    print(news_src)

    #Getting word/sentence in the link above
    news_id = news_src.split('/')[3]
    print("ID after website link is :",news_id)

    #Make link to webpage from id
    web_link = f'https://thestandard.co/{news_id}'
    print("Your generated weblink : ", web_link, "\n")
    print()
  except Exception as e:
    #Pass to SKIP
    #pass

    #Still want but set to none
    web_link= none
```

```
#Insert data to CSV file

import csv

csv_file = open('scraping.csv','w')

csv_writer = csv.writer(csv_file)
csv_writer.writerow(['headline','link_id','link'])


for article in soup.find_all("h2",class_='news-title'):
  headline = article.a.text
  print("Today News :",headline)
  try:
    #Getting link to the news (inside tag)
    news_src = article.find('a')['href']
    print(news_src)
    #Getting word/sentence in the link above
    news_id = news_src.split('/')[3]
    print("ID after website link is :",news_id)
    #Make link to webpage from id
    web_link = f'https://thestandard.co/{news_id}'
    print("Your generated weblink : ", web_link, "\n")
    print()
  except Exception as e:
    web_link= none

 
  csv_writer.writerow([headline, news_id, web_link])
  
csv_file.close()
```
